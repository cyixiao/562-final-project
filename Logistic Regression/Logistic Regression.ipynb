```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Load data
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
print (x_train)
```

```python
# Normalize data
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
# Flatten images
nsamples, nx, ny, nrgb = x_train.shape
x_train_flat = x_train.reshape((nsamples, nx*ny*nrgb))
x_test_flat = x_test.reshape((x_test.shape[0], nx*ny*nrgb))
```

```python
# Convert labels to one-hot encoding
y_train_cat = to_categorical(y_train, num_classes=10)
y_test_cat = to_categorical(y_test, num_classes=10)
```

```python
# Build logistic regression model
model = Sequential([
    Dense(10, input_dim=nx*ny*nrgb, activation='softmax')
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
early_stop = EarlyStopping(monitor='val_loss', patience=5)
model.fit(x_train_flat, y_train_cat, epochs=50, batch_size=32, validation_split=0.1, callbacks=[early_stop])
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(x_test_flat, y_test_cat)
print(f"Test accuracy: {accuracy:.3f}")
```

```python
# Predictions for further evaluation
y_pred = model.predict(x_test_flat)
y_pred_classes = np.argmax(y_pred, axis=1)

print(classification_report(y_test, y_pred_classes))
print(confusion_matrix(y_test.ravel(), y_pred_classes))
```
